## **Visualizations Based on the Transfromed Data**

### **Full Data**

```python
# Import pandas for data manipulation and analysis
import pandas as pd

# Import matplotlib for data visualization (plotting graphs)
import matplotlib.pyplot as plt
```

```python
# Reading the CSV file named 'transformed_full.csv' into a DataFrame called df_full
df_full = pd.read_csv("transformed_full.csv")
```

```python
# üìä Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# üí∞ Group data to get total sales per region, sorted in descending order
region_sales = df_full.groupby('region')['total_price'].sum().sort_values(ascending=False)

# üñºÔ∏è Set up the plot size
plt.figure(figsize=(8, 5))

# üß± Create a horizontal barplot for total sales by region
sns.barplot(x=region_sales.values, y=region_sales.index, color='skyblue')

# üè∑Ô∏è Add plot title and labels
plt.title('Total Sales by Region')
plt.xlabel('Total Price')
plt.ylabel('Region')

# üé® Adjust layout for better spacing
plt.tight_layout()

# üì∑ Display the plot
plt.show()
```

### üìà Insight: Total Sales by Region

- ü•á **South** region had the **highest total sales**, indicating it may be the most profitable or has the most customer activity.
- ü•à **West** region followed as the second-highest contributor in total sales.
- ü•â **North** region came in third, suggesting moderate sales performance.
- üß≠ **East** region recorded the **lowest total sales**, indicating either lower demand, fewer transactions, or potential for growth.

üîé This distribution could inform strategic decisions such as:
- Increasing marketing efforts in the East.
- Create additional branch locations in the South.
- Investigating why the South performs better‚Äîproduct mix, customer base, etc.
